```python
df['genres']=df['genres'].fillna('NaN')
```

```python
df_index=df[df.genres == 'NaN'].index.tolist() #get the index
df=df.drop(df_index) #delete the index
df.shape
```

```python
df=df.drop('genres',axis=1).join(df['genres'].str.split('|',expand=True).stack().reset_index(level=1,drop=True).rename('genres'))
df.head(3)
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df.shape
```

```python
df=df[df['budget_adj']>0]
df=df[df['popularity']>0]
```

```python
df.info()
```

```python
df['genres'].value_counts()
```

```python
budget=df.groupby('genres')['budget_adj'].mean().sort_values(ascending=False)
```

```python
budget
```

```python
budget.plot(kind='bar',
           color='blue')
plt.ylabel('budget')
plt.title('budget of different genres')
```

```python
popularity= df.groupby(['genres'])['popularity'].mean().sort_values(ascending=False) #arrange popularity descending
```

```python
most_popularity=popularity[:5].index.tolist() #select top 5
most_popularity
```

```python
df_popularity=df[['genres','release_year','popularity']].set_index('genres').loc[most_popularity].reset_index('genres')
df_popularity_year=pd.pivot_table(df_popularity,values='popularity',index='release_year',columns='genres')
```

```python
df_popularity_year.plot(kind='line',subplots=True,sharex=True,sharey=True,figsize=(20,20))
```

```python
df.groupby('genres')['vote_average'].mean().sort_values(ascending=False).index[0]
```
